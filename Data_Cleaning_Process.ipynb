```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

```python
#DATA CLEANING PROCESS
df = pd.read_csv("/content/Datasci_salary/glassdoor_jobs.csv")
df.head(100)
```

```python
df.shape
```

```python
#Salary parsing
#Company name text only
#age of company
#parsing of job description (Pythoin,etc)
```

```python
df =df[df["Salary Estimate"]!='-1']
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary.head()
```

```python
minus_kd=salary.apply(lambda x: x.replace('K','').replace('$',''))
minus_kd.head(10)
```

```python
df['hourly']=df["Salary Estimate"].apply(lambda x: 1 if 'per hour' in x.lower()else 0)

df.head()
```

```python
df['employer provoded salary']=df["Salary Estimate"].apply(lambda x: 1 if 'employer provided salary' in x.lower()else 0)
```

```python
df.head()
```

```python
min_hr = minus_kd.apply(lambda x: x.lower().replace("per hour", "").replace("employer provided salary:",""))
min_hr.head(100)
minus_kd.head(500)
```

```python
df['min_salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))
df['min_salary'].dtype

df['max_salary'] = min_hr.apply(lambda x: int(x.split('-')[1]))
df.head()
df['avg _salary'] = (df.min_salary+df.max_salary)/2
df.head()
```

```python
#Company Name
df['Company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-3],axis = 1)
```

```python
df.head(10)
```

```python
#State_Field
df['job_state']=df["Location"].apply(lambda x: x.split(',')[1])
df.job_state.value_counts()
     
```

```python
df.columns

```

```python
df["same_state"]=df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis =1)
```

```python
df.columns
```

```python
#Age of the Company
df['age']=df.Founded.apply(lambda x: x if x <1 else 2020 -x)
df[df['age']==df['age'].max()]
df['age'].max()
```

```python
#Sparsing Jobs description (Python, Etc.)
df['Job Description'][0]
```

```python
#Python, R Studio, Spark, AWS, Excel
df['Python_yn']=df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df.columns
```

```python
df['RStudio_yn']=df['Job Description'].apply(lambda x: 1 if 'r studio'in x.lower() or 'r-studio' in x.lower() else 0)
df.columns
```

```python
df.RStudio_yn.value_counts()
```

```python
df['Spark_yn']=df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df.Spark_yn.value_counts()
```

```python
df['aws_yn']=df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df.aws_yn.value_counts()
```

```python
df['excel_yn']=df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df.excel_yn.value_counts()
```

```python
df.columns
```

```python
#Drop first Column
df_drop = df.drop(['Unnamed: 0'], axis = 1)
df_drop.columns
```

```python
df_drop.to_csv('salary_data_cleaned.csv',index = False)
pd.read_csv('salary_data_cleaned.csv')
```
